```python
class Animal:
    def __init__(self, name):
        self.name = name
    def move(self):
        print(f"{self.name}가 이동")

class Dog(Animal):
    def bark(self):
        print(f"{self.name}가 멍멍")

a = Animal('abc')
a.move()
dog = Dog("초코")
dog.move()
dog.bark()
```
