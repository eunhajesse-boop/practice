```python
sentence = "[ first in ] ( first out )."
open_parenthesis = []
close_parenthesis = []
open_bracket = []
close_bracket = []

for char in sentence: #이렇게만 써도 문자열 하나씩 split됨
    if char == '(':
        open_parenthesis.append(char)
    if char == ')':
        close_parenthesis.append(char)
    if char == '[':
        open_bracket.append(char)
    if char == ']':
        close_bracket.append(char)



if sentence[-1] != '.':
    print('No')
elif len(open_parenthesis) == len(close_parenthesis) and len(open_bracket) == len(close_bracket):
    print('YES')
else:
    print('NO')


```
